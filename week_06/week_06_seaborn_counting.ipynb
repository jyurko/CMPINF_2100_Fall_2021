### Intro to Seaborn - counting 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.info()
```

```python
penguins.describe()
```

```python
penguins.species.value_counts()
```

```python
penguins.describe(include=[object])
```

```python
penguins.nunique()
```

```python
penguins.isna().sum()
```

Start counting the `species`.

```python
penguins.species.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

penguins.species.value_counts().plot(kind='bar', ax=ax)

plt.show()
```

```python
fig2, ax = plt.subplots(figsize=(11, 2))

penguins.species.value_counts().plot(kind='barh', ax=ax)

plt.show()
```

Seaborn will do the counting for us with the `sns.countplot()` is an axes-level function to create a bar chart.

```python
sns.countplot(data = penguins, x='species')

plt.show()
```

To make a horizontal bar chart we use the `y` variable instead of `x` in `sns.countplot()`.

```python
sns.countplot(data = penguins, y='species')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(11, 4))

sns.countplot(data = penguins, y='species', ax=ax)

plt.show()
```

To make a bar chart we can also use the **figure level** function `sns.catplot()`.

```python
sns.catplot(data = penguins, y='species', kind='count')

plt.show()
```

We can control the facets (subplots) with the `col` and `row` argument to figure level Seaborn functions.

```python
sns.catplot(data = penguins, x='species', kind='count', col='island')

plt.show()
```

Create a **dodged bar chart** where the side-by-side bars of different color denote the values of another variable.

```python
sns.catplot(data = penguins, x='species', kind='count', col='island', hue='sex')

plt.show()
```

Check these numbers make sense by grouping and aggregating.

```python
penguins.groupby(['island', 'species', 'sex']).size().reset_index(name='num_rows')
```

Heat maps are another option for displaying counts of combinations. I like to make heat maps from **cross-tabulations**.

For example, let's get the number of rows for each combination of island and species.

```python
penguins.groupby(['island', 'species']).size().reset_index(name='num_rows')
```

```python
pd.crosstab( penguins.island, penguins.species )
```

Visualize the cross-tabulation as a heatmap.

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(pd.crosstab( penguins.island, penguins.species ), ax=ax )

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(pd.crosstab( penguins.island, penguins.species ), ax=ax,
            annot=True, annot_kws={'size': 30}, cbar=False)

plt.show()
```

## Histogram

Make a histogram for the `flipper_length_mm` variable.

```python
sns.histplot(data = penguins, x='flipper_length_mm')

plt.show()
```

```python
sns.histplot(data = penguins, x='flipper_length_mm', bins=4)

plt.show()
```

```python
sns.histplot(data = penguins, x='flipper_length_mm', bins=101)

plt.show()
```

```python
sns.histplot(data = penguins, x='flipper_length_mm', bins=225)

plt.show()
```

Make a categorical variable out of the continuous variable.

```python
penguins_b = penguins.copy()
```

```python
penguins_b['flipper_cat'] = pd.cut( penguins_b.flipper_length_mm, bins=4, include_lowest=True, ordered=True )
```

```python
penguins_b.flipper_cat.value_counts()
```

```python
sns.catplot(data = penguins_b, x='flipper_cat', kind='count')

plt.show()
```

```python
sns.histplot(data = penguins, x = 'flipper_length_mm', bins=4)

plt.show()
```

```python
sns.kdeplot(data = penguins, x='flipper_length_mm')

plt.show()
```

Use colors to denote the grouping by a categorical variable.

```python
sns.kdeplot(data = penguins, x='flipper_length_mm', hue='species')

plt.show()
```

```python
sns.kdeplot(data = penguins, x='flipper_length_mm', hue='species', common_norm=False)

plt.show()
```

```python
sns.displot(data = penguins, x='flipper_length_mm', kde=True, rug=True)

plt.show()
```

### Boxplots to summarize a continuous variable by a categorical

```python
sns.catplot(data = penguins, x='species', y='flipper_length_mm')

plt.show()
```

```python
sns.catplot(data = penguins, x='species', y='flipper_length_mm', kind='box')

plt.show()
```

```python
sns.catplot(data = penguins, x='species', y='flipper_length_mm', kind='box',
            showmeans=True,
            meanprops={'marker': 'o', 'markerfacecolor': 'white', 'markeredgecolor': 'black'})

plt.show()
```

```python
sns.catplot(data = penguins, x='species', y='flipper_length_mm', kind='box', hue='sex',
            showmeans=True,
            meanprops={'marker': 'o', 'markerfacecolor': 'white', 'markeredgecolor': 'black'})

plt.show()
```

```python
sns.pointplot(data = penguins, x='species', y='flipper_length_mm', ci=95)

plt.show()
```
