```python
plt.figure()

### create a subplot

plt.subplot(1, 1, 1) ### (number of rows, number of columns, current plot index - panel number)

### plot a sine wave
plt.plot(np.sin(np.linspace(0, 10, 100)))

### show the figure
plt.show()
```

One figure with two subplots. But only make one of the subplots.

```python
plt.figure()

### create the first subplot on a grid of 2 rows and 1 column
plt.subplot(2, 1, 1)

### plot the data
plt.plot(np.sin(np.linspace(0, 10, 100)))

### show the plot

plt.show()
```

Now make both subplots.

```python
plt.figure()

### create the first subplot on a grid of 2 rows and 1 column
plt.subplot(2, 1, 1)

### plot the data
plt.plot(np.sin(np.linspace(0, 10, 100)))

### create the second subplot on a grid of 2 rows and 1 column
plt.subplot(2, 1, 2)

### plot the data on the second subplot
plt.plot(np.cos(np.linspace(0, 10, 100)))

### show the plot

plt.show()
```

## Object Oriented style

* step 1: define the figure and axes objects

* step 2: plot the data using the methods associated with the axes objects

* step 3: modify axes object attributes (if needed)

* step 4: modify figure object attributes (if needed)

* step 5: show the figure

Make a figure with 2 subplots - or 2 axes on the figure.

```python
fig, ax = plt.subplots(2)
```

```python
%whos
```

```python
print( type(ax) )
```

```python
print( type(ax[0]) )
```

```python
print( type(ax[1]) )
```

Plot the sin and cos waves on the individual axes.

```python
ax[0].plot(np.sin(np.linspace(0, 10, 100)))
```

```python
ax[1].plot(np.cos(np.linspace(0, 10, 100)))
```

Create a grid a of subplots.

```python
fig, ax = plt.subplots(2, 3, figsize=(12, 6), sharex=True, sharey=True)
```

```python
print( type(ax) )
```

```python
print( ax.shape )
```

```python
print( type(ax[0, 1]) )
```

```python
for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i, j].text(0.5, 0.5, str((i,j)), ha='center', fontsize=20)
```
