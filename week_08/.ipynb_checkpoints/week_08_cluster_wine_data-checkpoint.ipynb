```python
wine_data = pd.read_csv( wine_url, names=wine_names )
```

```python
wine_data.Cultivar.value_counts()
```

```python
wine_data['Cultivar'] = wine_data.Cultivar.astype('category')
```

## Preprocess

```python
from sklearn.preprocessing import StandardScaler
```

```python
wine_features = wine_data.select_dtypes('number').copy()
```

```python
Xwine = StandardScaler().fit_transform( wine_features )
```

If we do not standardize these variables...

```python
sns.catplot(data = wine_data, kind='box', aspect=3)

plt.show()
```

The standardized features are all roughly the same scale.

```python
sns.catplot(data = pd.DataFrame( Xwine, columns=wine_features.columns), kind='box', aspect=3)

plt.show()
```

## PCA

We will use PCA, specifically the first two PCs to streamline the visualizations.

```python
from sklearn.decomposition import PCA
```

```python
wine_pcs = PCA(n_components=2).fit_transform( Xwine )
```

```python
wine_pcs.shape
```

```python
wine_pcs_df = pd.DataFrame( wine_pcs, columns=['pc_01', 'pc_02'])
```

```python
wine_pcs_df['Cultivar'] = wine_data.Cultivar
```

Visualize the PC scores as a scatter plot.

```python
sns.set_style('whitegrid')

sns.relplot(data = wine_pcs_df, x='pc_01', y='pc_02', s=100, height=8)

plt.show()
```

Color the PC score scatter plot by `Cultivar`.

```python
sns.relplot(data = wine_pcs_df, x='pc_01', y='pc_02', hue='Cultivar', s=100, height=8)

plt.show()
```

How would the observations group together if we only focus on the numeric features?

## KMeans

```python
from sklearn.cluster import KMeans
```

Cluster the observations using the standardized continuous features into 2 groups.

```python
cluster_2 = KMeans(n_clusters=2, random_state=121, n_init=25, max_iter=500).fit_predict( Xwine )
```

```python
wine_pcs_df['k2'] = pd.Series( cluster_2, index=wine_pcs_df.index)

wine_pcs_df['k2'] = wine_pcs_df.k2.astype('category')
```

```python
wine_pcs_df.k2.value_counts()
```

```python
sns.relplot(data = wine_pcs_df, x='pc_01', y='pc_02', hue='k2', s=100, height=8)

plt.show()
```

Calculate the total within sum of squares for 1 cluster up to 30 clusters.

```python
tots_within = []

K = range(1, 31)

for k in K:
    km = KMeans(n_clusters=k, random_state=121, n_init=25, max_iter=500).fit( Xwine )
    tots_within.append( km.inertia_ )
```

```python
print( len(tots_within) )
```

```python
print( len(list(K)) )
```

Visualize the results.

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(K, tots_within, 'o-')
ax.set_xlabel('number of clusters')
ax.set_ylabel('total within sum of squares')

plt.show()
```

Calculate the silhouette coefficient to identify the optimal number of clusters.

```python
from sklearn.metrics import silhouette_score
```

Calculate the silhouette coefficient for 2 to 30 clusters.

```python
sil_coef = []

K = range(2, 31)

for k in K:
    k_label = KMeans(n_clusters=k, random_state=121, n_init=25, max_iter=500).fit_predict( Xwine )
    sil_coef.append( silhouette_score(Xwine, k_label) )
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(K, sil_coef, 'o-')
ax.set_xlabel('number of clusters')
ax.set_ylabel('average silhouette coefficient')

plt.show()
```

## Hierarchical clustering

```python
from scipy.cluster import hierarchy
```

Pros and cons of linkages.

https://scikit-learn.org/stable/auto_examples/cluster/plot_linkage_comparison.html#sphx-glr-auto-examples-cluster-plot-linkage-comparison-py

### Complete linkage

```python
hclust_complete = hierarchy.complete( Xwine )
```

```python
print( type(hclust_complete) )
```

Visualize the results as a **dendrogram**.

```python
fig = plt.figure(figsize=(16,8))

dn = hierarchy.dendrogram( hclust_complete )

plt.show()
```

Remove the x-axis labels.

```python
fig = plt.figure(figsize=(16,8))

dn = hierarchy.dendrogram( hclust_complete, no_labels=True )

plt.show()
```

### Single linkage

```python
hclust_single = hierarchy.single( Xwine )
```

```python
fig = plt.figure(figsize=(16,8))

dn = hierarchy.dendrogram( hclust_single, no_labels=True )

plt.show()
```

### Average linkage

```python
hclust_average = hierarchy.average( Xwine )
```

```python
fig = plt.figure(figsize=(16,8))

dn = hierarchy.dendrogram( hclust_average, no_labels=True )

plt.show()
```

### Centroid linkage

```python
hclust_centroid = hierarchy.centroid( Xwine )
```

```python
fig = plt.figure(figsize=(16,8))

dn = hierarchy.dendrogram( hclust_centroid, no_labels=True )

plt.show()
```

### Ward method

```python
hclust_ward = hierarchy.ward( Xwine )
```

```python
fig = plt.figure(figsize=(16,8))

dn = hierarchy.dendrogram( hclust_ward, no_labels=True )

plt.show()
```

### Cut the tree

```python
ward_cut_3 = hierarchy.cut_tree( hclust_ward, n_clusters=3 )
```

```python
print( type(ward_cut_3) )
```

```python
print( ward_cut_3.shape )
```

```python
print( ward_cut_3.ndim )
```

```python
ward_cut_3.ravel()
```

```python
print( ward_cut_3.ravel().ndim )
```

```python
wine_pcs_df['hclust_3'] = pd.Series( ward_cut_3.ravel(), index=wine_pcs_df.index)

wine_pcs_df['hclust_3'] = wine_pcs_df.hclust_3.astype('category')
```

```python
sns.relplot(data = wine_pcs_df, x='pc_01', y='pc_02', hue='hclust_3', s=100, height=8)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,7))

sns.heatmap(pd.crosstab(wine_pcs_df.hclust_3, wine_pcs_df.Cultivar, margins=True),
           annot=True, annot_kws={'size': 25})

plt.show()
```
