## Fit logistic regression models

```python
import statsmodels.formula.api as smf
```

Start simple fit a model with all linear additive features with just the continuous inputs.

$$\log\left(\frac{\mathrm{Probability}_n}{1 - \mathrm{Probability}_n} \right) = \beta_0 + \beta_1 x_{n,1} + \beta_2 x_{n,2} + \beta_3 x_{n,3} + \beta_4 x_{n,4} $$

```python
fit_01 = smf.logit(formula = 'y ~ x1 + x2 + x3 + x4', data = df).fit()
```

```python
print( fit_01.summary() )
```

I prefer to visualize the coefficient summaries.

```python
def my_coefplot(model_object, figsize_use=(10,5)):
    fig, ax = plt.subplots(figsize=figsize_use)
    
    ax.errorbar(y = model_object.params.index,
                x = model_object.params,
                xerr = 2 * model_object.bse,
                fmt='o', color='black', ecolor='black',
                elinewidth=3, ms=10)
    
    ax.axvline(x=0, linestyle='--', linewidth=5, color='grey')
    
    ax.set_xlabel('coefficient value')
    
    plt.show()
```

```python
my_coefplot(fit_01)
```

All pair-wise interactions between the continuous inputs.

```python
fit_02 = smf.logit(formula = 'y ~ (x1 + x2 + x3 + x4) ** 2', data = df).fit()
```

If we want to know the corresponding probability we need to convert from the log-odds value to the probability value.

```python
np.exp(-0.5) / (1 + np.exp(-0.5))
```

Calculate the value associated with the `x5==c` dummy variable.

```python
np.exp(-0.5 + 0.17) / ( 1 + np.exp(-0.5 + 0.17))
```

The dummy variable represents the **relative effect** from the reference level.

$$ \beta_0 + \beta_{x5b} x_{n, x5b} + \beta_{x5c} x_{, x5c} $$

When `x5 == 'c'`, the expression simplifies to:

$$ \beta_0 + \beta_{x5b} \times 0 + \beta_{x5c} \times 1$$

$$ \beta_0 + \beta_{x5c} $$

## Binary classification performance metrics

```python
mod_list = [fit_01, fit_02, fit_03, fit_04, fit_05, fit_06, fit_07, fit_08, fit_09, fit_10]
```

```python
print( mod_list[2].summary() )
```

Predict the event probability on the training set for each model.

```python
df_copy_b = df.copy()
```

```python
'pred_probability_'+str(1)
```

```python
'pred_probability_'+str(10)
```

```python
'pred_probability_'+str(1).zfill(2)
```

```python
for i, mod in enumerate(mod_list):
    df_copy_b['pred_probability_'+str(i+1).zfill(2)] = mod.predict(df)
```

```python
df_copy_b
```

Classify by comparing the predicted probability to a threshold. Use the default threshold of 0.5.

```python
for i in range(len(mod_list)):
    df_copy_b['pred_class_'+str(i+1).zfill(2)] = np.where( df_copy_b['pred_probability_'+str(i+1).zfill(2)] > 0.5, 1, 0)
```

```python
df_copy_b
```

Calculate the accuracy on the training set.

```python
model_accuracy = []

for i in range(len(mod_list)):
    model_accuracy.append( df_copy_b.loc[ df_copy_b.y == df_copy_b['pred_class_'+str(i+1).zfill(2)] ].shape[0] / df_copy_b.shape[0])
```

```python
model_accuracy
```

### ROC curve

```python
from sklearn.metrics import roc_curve
```

```python
def roc_values(mod_id, df_object):
    fpr, tpr, threshold = roc_curve(df_object.y.to_numpy(), df_object['pred_probability_'+str(mod_id+1).zfill(2)].to_numpy())
    res = pd.DataFrame({'fpr': fpr, 'tpr': tpr, 'threshold': threshold})
    res['model_name'] = 'fit_'+str(mod_id+1).zfill(2)
    
    return res
```

```python
all_roc_curves = []

for i in range(len(mod_list)):
    all_roc_curves.append( roc_values(i, df_copy_b) )
```

```python
all_roc_df = pd.concat(all_roc_curves).reset_index()
```

```python
all_roc_df
```

```python
sns.relplot(data = all_roc_df, x='fpr', y='tpr', hue='model_name',
            estimator=None, units='model_name',
            kind='line',
            drawstyle='steps-post',
            height=9)

plt.show()
```
