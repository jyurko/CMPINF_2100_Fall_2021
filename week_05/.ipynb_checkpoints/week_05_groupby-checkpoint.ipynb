### Apply custom functions to group

Create out own function which calculates the IQR.

```python
def my_iqr(values):
    q25 = np.percentile(values, q=25)
    q75 = np.percentile(values, q=75)
    
    return q75 - q25
```

We need to combine `.groupby()` with the `.aggregate()` or the `.agg()` function.

```python
gap_df.groupby('year').lifeExp.aggregate(my_iqr)
```

```python
gap_year_summary = gap_df.groupby('year').lifeExp.describe()
```

```python
gap_year_summary
```

```python
gap_year_summary['IQR'] = gap_year_summary['75%'] - gap_year_summary['25%']
```

```python
gap_year_summary
```

```python
gap_year_summary.IQR
```

```python
gap_df.groupby('year').lifeExp.aggregate(my_iqr)
```

```python
gap_df.groupby('year').lifeExp.mean()
```

```python
gap_df.groupby('year').lifeExp.sem()
```

Apply multiple functions per group.

```python
gap_df.groupby('year').lifeExp.aggregate( [ 'mean', 'sem' ])
```

```python
gap_df.groupby('year').lifeExp.aggregate( [ np.min, np.max, 'mean', 'sem' ])
```

```python
gap_df.groupby('year').lifeExp.aggregate( [ np.min, np.max, my_iqr, 'mean', 'sem' ])
```

We can rename columns after performing the groupby and aggregation.

```python
gap_df.groupby('year')['lifeExp'].\
aggregate(['mean', 'min', 'max', 'sem']).\
rename(columns={'mean': 'avg_lifeExp',
                'min': 'min_lifeExp',
                'max': 'max_lifeExp', 
                'sem': 'se_avg'})
```

We can reset index to bring the grouping variable back into the DataFrame.

```python
gap_df.groupby('year')['lifeExp'].\
aggregate(['mean', 'min', 'max', 'sem']).\
rename(columns={'mean': 'avg_lifeExp',
                'min': 'min_lifeExp',
                'max': 'max_lifeExp', 
                'sem': 'se_avg'}).\
reset_index()
```

### Group by multiple columns

```python
gap_df.groupby(['year', 'continent']).size()
```

THe index is a multi-index with the hierarchical structuring of the rows.

```python
gap_df.groupby(['year', 'continent']).lifeExp.mean()
```

If we don't specify the column the method is applied to all columns.

```python
gap_df.groupby(['year', 'continent']).mean()
```

Reset to remove the multi-index.

```python
gap_df.groupby(['year', 'continent']).mean().reset_index()
```

### Apply a specific function to specific columns

Use the `pd.NamedAgg()` function to control this behavior.

```python
gap_df.groupby(['year', 'continent']).\
aggregate(lifeExp_avg = pd.NamedAgg(column='lifeExp', aggfunc='mean'),
          lifeExp_sem = pd.NamedAgg(column='lifeExp', aggfunc='sem'),
          num_countries = pd.NamedAgg(column='country', aggfunc='nunique'),
          gdpPercap_avg = pd.NamedAgg(column='gdpPercap', aggfunc=np.mean),
          gdpPercap_iqr = pd.NamedAgg(column='gdpPercap', aggfunc=my_iqr))
```

We can use a shortcut method with tuples to do everything that `pd.NamedAgg()` does.

```python
gap_df.groupby(['year', 'continent']).\
aggregate(lifeExp_avg = ('lifeExp', 'mean'),
          lifeExp_sem = ('lifeExp', 'sem'),
          num_countries = ('country', 'nunique'),
          gdpPercap_iqr = ('gdpPercap', my_iqr))
```

```python
gap_df.groupby(['year', 'continent']).\
aggregate(lifeExp_avg = ('lifeExp', 'mean'),
          lifeExp_sem = ('lifeExp', 'sem'),
          num_countries = ('country', 'nunique'),
          gdpPercap_iqr = ('gdpPercap', my_iqr)).\
reset_index()
```

```python
gap_df.groupby(['year', 'continent']).\
aggregate(lifeExp_avg = ('lifeExp', 'mean'),
          lifeExp_sem = ('lifeExp', 'sem'),
          num_countries = ('country', 'nunique'),
          gdpPercap_iqr = ('gdpPercap', my_iqr),
          num_rows = ('country', 'size')).\
reset_index()
```

Save the summarized data set with the multi-index.

```python
g = gap_df.groupby(['year', 'continent']).\
aggregate(lifeExp_avg = ('lifeExp', 'mean'),
          lifeExp_sem = ('lifeExp', 'sem'),
          num_countries = ('country', 'nunique'),
          gdpPercap_iqr = ('gdpPercap', my_iqr),
          num_rows = ('country', 'size'))
```

```python
g
```

The multi-index exists as a tuple.

```python
g.index
```

However, there is a hierarchy. We do not have to enter the tuple if we wish to find locations based on the primary index.

```python
g.loc[1952]
```

However, if we try to select rows based on the secondary index alone...we get an error!

```python
g.loc['Asia']
```

We need the tuple when we want to select the complete row based on the index value.

```python
g.loc[(1952, 'Asia')]
```

However, the `.iloc[]` attribute still works since it is based on the position and not the value.

```python
g.iloc[0]
```
