If a column in the data set corresponds to our desired index, we can specify that with the `index_col` argument.

```python
dfA = pd.read_excel('small_example.xlsx', index_col=0)
```

```python
dfA
```

```python
dfA.info()
```

```python
dfD = pd.read_excel('small_example.xlsx', index_col=3)
```

```python
dfD
```

```python
print( df0.index )

print( dfA.index )

print( dfD.index )
```
