### Reading gapminder data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Set the URL path to the data.

```python
gap_url = 'https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv'
```

```python
gap_df = pd.read_csv(gap_url, sep = '\t')
```

```python
gap_df.info()
```

```python
gap_df.describe()
```

```python
gap_df.describe(include='all')
```

```python
gap_df.describe(include=[object])
```

What are the unique values?

```python
gap_df.continent.value_counts()
```

```python
gap_df
```

Visualize the counts with a bar chart.

```python
fig, ax = plt.subplots(figsize=(12,6))

gap_df.continent.value_counts().plot(kind='bar', ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

gap_df.continent.value_counts().plot(kind='barh', ax=ax)

plt.show()
```

```python
gap_df.continent.value_counts(normalize=True)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

gap_df.continent.value_counts(normalize=True).plot(kind='barh', ax=ax)

plt.show()
```

## Merge a smaller data set

```python
my_travel_df = pd.DataFrame({'country': ['United States', 'Canada', 'United Kingdom', 'Czech Republic', 'My backyard'],
                             'status': ['live there', 'visited', 'visited', 'visited', 'live there']})
```

```python
my_travel_df
```

```python
gap_df
```

```python
pd.merge( gap_df, my_travel_df, on='country')
```

```python
pd.merge( gap_df, my_travel_df, on='country').shape
```

```python
pd.merge( gap_df, my_travel_df, on='country', how='left')
```

```python
pd.merge( gap_df, my_travel_df, on='country', how='right')
```
