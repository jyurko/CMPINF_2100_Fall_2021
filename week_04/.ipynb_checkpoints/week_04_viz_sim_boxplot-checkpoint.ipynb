```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rg = np.random.default_rng(12345)
```

```python
X = rg.random((50, 7))
```

```python
X.shape
```

```python
type(X)
```

```python
Xdf = pd.DataFrame(X)
```

```python
Xdf.info()
```

```python
Xdf
```

```python
Xdf.describe()
```

```python
Xdf.T
```

```python
Xdf.T.describe()
```

```python
fig, ax = plt.subplots(figsize=(18, 6))

Xdf.T.boxplot(showmeans=True)

plt.show()
```

```python
X700df = pd.DataFrame(rg.random((50,700)))
```
