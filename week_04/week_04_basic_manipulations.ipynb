## Conditionally Select Rows or FILTER

Find all rows where `B` is even.

```python
df
```

```python
df.loc[ df['B']%2 == 0 ]
```

```python
df['B']%2 == 0
```

```python
df.loc[ df.B%2 == 0 ]
```

The conditional filter does NOT modify in place.

```python
df
```

```python
df_even = df.loc[ df.B%2 == 0 ]
```

```python
df_even
```

```python
df_small = df.loc[ df.B%2 == 0].copy().reset_index(drop=True)
```

```python
df_small
```

We can filter based on the values of strings.

```python
df_small.loc[ df_small['E'] == 'cc' ]
```

We can match based on multiple conditions.

```python
df_small.loc[ (df_small.E == 'cc') | (df_small.E == 'bb') ]
```

The OR operation can be streamlined with the `.isin()` method.

```python
df_small.loc[ df_small.E.isin(['cc', 'bb']) ]
```

```python
df_small.loc[ df_small.B.isin([0, 6, 10]) ]
```

We can filter and select a subset of columns.

```python
df_small.loc[ df_small.B.isin([0, 6, 10]), ['A', 'B', 'E']]
```

```python
df_small.loc[ df_small.B.isin([0, 6, 10]), : ]
```
